```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

checkpoint = "CohereForAI/aya-101"

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
aya_model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
```

```python
input_01 = tokenizer.encode("Translate to English: Жирафите са високи животни.", return_tensors="pt")
output_01 = aya_model.generate(input_01, max_new_tokens=128)

print(tokenizer.decode(output_01[0]))
```

```python
input_02 = tokenizer.encode("Translate to Bulgarian: Giraffes are tall animals.", return_tensors="pt")
output_02 = aya_model.generate(input_02, max_new_tokens=128)

print(tokenizer.decode(output_02[0]))
```
