```python
from transformers import pipeline

extractor = pipeline(
    'feature-extraction',
    model='antoinelouis/colbert-xm'
)
```

```python
output = extractor(
    'This is a simple test.',
    return_numpy=True
)
```

```python
print(len(output[0]))
print(len(output[0][0]))
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('antoinelouis/colbert-xm')
```

```python
import numpy as np

tokenized_input = tokenizer(
    'This is a simple test.',
    return_tensors='np'
)

token_ids = tokenized_input['input_ids'][0]
print(len(token_ids))

unused_tokens = set(
    [
        tokenizer.cls_token_id,
        tokenizer.eos_token_id,
        tokenizer.pad_token_id,
        tokenizer.unk_token_id,
    ]
)

colbert_embeddings = output[0]
print(len(colbert_embeddings))

for token_id, colbert_embedding in zip(
    token_ids,
    colbert_embeddings
):
    if token_id not in unused_tokens:
        colbert_token_item = {}
        colbert_token_item[token_id] = colbert_embedding

        print(colbert_token_item)
```
