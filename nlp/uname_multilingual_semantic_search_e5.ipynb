```python
import numpy as np
from sentence_transformers import SentenceTransformer
from usearch.index import Index
```

```python
encoder = SentenceTransformer("intfloat/multilingual-e5-large")
```

```python
index = Index(ndim=encoder.get_sentence_embedding_dimension())
```

```python
index.add(
    1,
    encoder.encode("Рим е столица на Италия.")
)

index.add(
    2,
    encoder.encode("Rome is the capital city of Italy.")
)

index.add(
    3,
    encoder.encode("Овчарските кучета са добри пазачи.")
)
```

```python
index_matches = index.search(encoder.encode("столица"), 10)
```

```python
for index_match in index_matches:
  print(f"{index_match.key} - {index_match.distance}")
```
