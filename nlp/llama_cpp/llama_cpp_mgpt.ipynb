```python
from llama_cpp import Llama
from huggingface_hub import hf_hub_download
```

```python
hf_hub_download(
    repo_id="Sosnitskij/mGPT-13B-gguf",
    filename="mGPT-13B-Q5_K_M.gguf",
    repo_type="model",
    local_dir="/tmp"
)
```

```python
mgpt_model = Llama(model_path="/tmp/mGPT-13B-Q5_K_M.gguf")

prompt = "Translate to Bulgarian: Giraffes are tall animals."

max_tokens = 100
temperature = 0.3
top_p = 0.1
echo = False
stop = ["Q", "\n"]

model_output = mgpt_model(
    prompt,
    max_tokens=max_tokens,
    temperature=temperature,
    top_p=top_p,
    echo=echo,
    stop=stop
)

final_result = model_output["choices"][0]["text"].strip()

# 2m
```
