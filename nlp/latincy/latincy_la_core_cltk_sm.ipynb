```python
import spacy

text = 'Dis Manibus Cai Cassi Felicis vixit annis LXX et Cassiae Lacaenae coniugi eius vixit annis LX et Vetsidio Supero filio piissimo qui vixit annis XII Hic siti sunt Cassia Cogitata parentibus et filio faciendum curavit'
```

```python
la_core_cltk_nlp = spacy.load('la_core_cltk_sm')

la_core_cltk_document = la_core_cltk_nlp(text)

for form in la_core_cltk_document:
  print(f'{form.text} -- {form.lemma_}')
```
