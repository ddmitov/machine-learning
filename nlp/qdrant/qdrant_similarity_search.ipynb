```python
import nltk
from nltk.tokenize import sent_tokenize

from qdrant_client import models, QdrantClient
from sentence_transformers import SentenceTransformer
```

```python
nltk.download('punkt')
```

```python
encoder = SentenceTransformer('silencesys/paraphrase-xlm-r-multilingual-v1-fine-tuned-for-medieval-latin')
```

```python
qdrant = QdrantClient(
    url='https://xxx.us-east-1-0.aws.cloud.qdrant.io:6333',
    api_key='api_key',
)
```

```python
search_phrase = 'exemplum'

hits = qdrant.search(
    collection_name='base',
    query_vector=encoder.encode(search_phrase).tolist(),
    limit=3
)

display('==============================')
display('Search Phrase:')
display(search_phrase)
display('==============================')

for hit in hits:
  display(
      hit.payload['name'],
      hit.payload['text'],
      'score:',
      hit.score
  )
  display('==========')
```
