```python
from colbert.modeling.checkpoint import Checkpoint
from colbert.infra import ColBERTConfig

ckpt = Checkpoint(
    "antoinelouis/colbert-xm",
    colbert_config=ColBERTConfig(root="experiments")
)
```

```python
vectors = ckpt.queryFromText(
    ["What does ColBERT do?", "Tis is a test."],
    bsize=16
)
```

```python
print(vectors)
```

```python
print(f"ColBERT Vectors: {len(vectors[0])}")
```

```python
print(f"ColBERT Vector Length: {len(vectors[0][0])}")
```
